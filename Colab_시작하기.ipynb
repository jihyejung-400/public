```
import pandas as pd

file_path = '/content/box3_test.xlsx'
df = pd.read_excel(file_path)
print(df.head())
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure and axes for the plot
plt.figure(figsize=(8, 6))

# Create a box plot to compare 'day 7' and 'day 14' data
sns.boxplot(data=df[['day 7', 'day 14']])

# Add title and labels with increased font size
plt.title('Distribution of Day 7 and Day 14 Data', fontsize=16) # Increased title font size
plt.ylabel('Value', fontsize=20) # Increased ylabel font size
plt.xlabel('Day', fontsize=20) # Increased xlabel font size
plt.xticks(fontsize=12) # Increased x-axis tick font size
plt.yticks(fontsize=12) # Increased y-axis tick font size

# Display the plot
plt.show()
```
