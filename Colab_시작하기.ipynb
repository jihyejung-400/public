```
import pandas as pd

file_path = '/content/box3_test.xlsx'
df = pd.read_excel(file_path)
```

```
print('First 5 rows of the DataFrame:')
display(df.head())
```

```
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# --- ANOVA Analysis ---
print("\n--- ANOVA Analysis ---")

treatment_groups = df['Treatment'].unique()

# Perform ANOVA for 'day 7'
day7_groups = [df[df['Treatment'] == group]['day 7'].dropna() for group in treatment_groups]
f_stat_day7, p_val_day7 = stats.f_oneway(*day7_groups)
print(f"ANOVA for 'day 7': F-statistic = {f_stat_day7:.3f}, P-value = {p_val_day7:.3f}")

# Perform ANOVA for 'day 14'
day14_groups = [df[df['Treatment'] == group]['day 14'].dropna() for group in treatment_groups]
f_stat_day14, p_val_day14 = stats.f_oneway(*day14_groups)
print(f"ANOVA for 'day 14': F-statistic = {f_stat_day14:.3f}, P-value = {p_val_day14:.3f}")

# --- Box Plots ---
print("\n--- Box Plots ---")

plt.figure(figsize=(14, 6))

# Box plot for 'day 7'
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
sns.boxplot(x='Treatment', y='day 7', data=df)
plt.title('Box Plot of Day 7 by Treatment')
plt.xlabel('Treatment')
plt.ylabel('Day 7 Value')
plt.xticks(rotation=45, ha='right')

# Box plot for 'day 14'
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
sns.boxplot(x='Treatment', y='day 14', data=df)
plt.title('Box Plot of Day 14 by Treatment')
plt.xlabel('Treatment')
plt.ylabel('Day 14 Value')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```
import plotly.express as px

# Interactive Box plot for 'day 7'
fig_day7 = px.box(df, x='Treatment', y='day 7',
                  title='Interactive Box Plot of Day 7 by Treatment',
                  labels={'Treatment': 'Treatment', 'day 7': 'Day 7 Value'})
fig_day7.update_xaxes(tickangle=45)
fig_day7.show()
```

```
import plotly.express as px

# Interactive Box plot for 'day 14'
fig_day14 = px.box(df, x='Treatment', y='day 14',
                   title='Interactive Box Plot of Day 14 by Treatment',
                   labels={'Treatment': 'Treatment', 'day 14': 'Day 14 Value'})
fig_day14.update_xaxes(tickangle=45)
fig_day14.show()
```
